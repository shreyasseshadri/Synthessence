```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
import pandas as pd
import os
```

```python
url="https://www.ebay.com/b/Laptops-Netbooks/175672/bn_1648276?_from=R40&_nkw=&_trksid=m570.l1313"
```

```python
driver = webdriver.Chrome('../Downloads/chromedriver')
```

```python
driver.get(url)
soup=BeautifulSoup(driver.page_source,"lxml")
```

```python
import pandas as pd
data=pd.DataFrame()
itemno=0
for item in soup.find_all("li",{'class':"s-item"}):
    try:
        data.loc[itemno,'title']=item.find_all("h3",{'class':"s-item__title"})[0].get_text()
        data.loc[itemno,'price']=item.find_all("span",{"class":"s-item__price"})[0].get_text()
        data.loc[itemno,'summary']=item.find_all("div",{'class':"s-item__summary"})[0].get_text()
    except:
        pass
    itemno+=1
```
